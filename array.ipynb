<b> - Valores Acima da Média <b>

```python
import numpy as np

aleatorio = [np.random.randint (0,100) for x in range(10)] #exercicio 14
aleatorio = np.array(aleatorio)
media = np.mean(aleatorio)
print("Média",media)
print("Valores maiores:",aleatorio[aleatorio > media])
```

<b> Análise de Notas <b>

```python
notas = np.array ([5, 7, 8, 6, 4, 9, 10]) #exercicio 15
media = np.mean(notas)
print(notas)
print("Média",media)
print("Valores maiores que a média:", notas[notas > media])
print(len(notas [notas > media])) #numeros acima da média geral
#sum (notas > media)
```
