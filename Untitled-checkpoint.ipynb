```python
def calcula_media_lista (lista):
    soma = sum(lista)
    quantidade = len(lista)
    media = soma/quantidade
    return media

print(calcula_media_lista([5,4,5,4] )) 
```

```python
def conta_vogais(texto) :
    contador = 0
    VOGAIS = ("A","E","I","O","U")
    texto = texto.upper()
    for letra in texto:
        if letra in VOGAIS:
            contador += 1
    return contador 

print (conta_vogais(input("Digite uma palavra para contar as vogais: ")))
```

```python
def imc (peso, altura):
    return peso / (altura ** 2)

peso =float(input("Digite seu peso (kg):"))
altura = float(input("Digite sua altura(m):"))

resultado = imc(peso, altura)
print(f"Seu IMC é : {resultado:.2f}")

```
